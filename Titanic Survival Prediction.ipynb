```python
import numpy as np
import tensorflow as tf
import pandas as pd
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
data_train1=pd.read_csv('train.csv')
data_test1=pd.read_csv('test.csv')
data_check1=pd.read_csv('test_result.csv')
```

```python
data_train1 = data_train1.drop(['Name','Ticket','Fare','Cabin','Embarked','Age'],axis =1)
data_test1 = data_test1.drop(['Name','Ticket','Fare','Cabin','Embarked','Age'], axis =1)
```

```python
data_train=data_train1.copy()
data_test=data_test1.copy()
```

```python
data_train = data_train[['PassengerId','Sex','SibSp','Parch','Pclass','Survived']]
data_test = data_test[['PassengerId','Sex','SibSp','Parch','Pclass']]
```

```python
data_train=data_train.values
data_test=data_test.values
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder_sex = LabelEncoder()
```

```python
data_train[:,1]  = label_encoder_sex.fit_transform(data_train[:,1])
data_test[:,1] = label_encoder_sex.fit_transform(data_test[:,1])
```

```python
data=data_train[:,0:5]
label=data_train[:,5]
data_test
```

```python
data = np.array(data, dtype=np.float64)
mean = data.mean(axis=0)
data -= mean
std = data.std(axis=0)
data //= std
data_test -= mean
data_test //= std
```

```python
#from sklearn.preprocessing import StandardScaler
#sc = StandardScaler()
#data = sc.fit_transform(data)
#X_test = sc.fit_transform(data_test)
```

```python
classifier= models.Sequential()
classifier.add(layers.Dense(3,  activation = 'relu', input_dim = 5))
#Hidden layer
classifier.add(layers.Dense(2, activation = 'relu'))
#output layer with 1 output neuron which will predict 1 or 0
classifier.add(layers.Dense(1, activation = 'sigmoid'))
```

```python
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier.fit(data, label, batch_size = 10,epochs = 100)
```

```python
#getting predictions of test data
prediction = classifier.predict(data_test).tolist()
# list to series
se = pd.Series(prediction)
# creating new column of predictions in data_check dataframe
data_check1['check'] = se
data_check1['check'] = data_check1['check'].str.get(0)
data_check1
```

```python
series = []
for val in data_check1.check:
    if val >= 0.5:
        series.append(1)
    else:
        series.append(0)
data_check1['final'] = series
data_check1
```
